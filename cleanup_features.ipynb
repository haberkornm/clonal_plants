# Aquatic feature

```python
wet = ['Sonchus oleraceus', 'Cyperus squarrosus', 'Echinochloa colona', 'Leptochloa viscida', 'Phalaris caroliniana', 
       'Polypogon monspeliensis', 'Zeltnera calycosa', 'Juncus balticus', 'Epilobium canum', 'Stemodia durantifolia', 
       'Polypogon viridis', 'Amaranthus crassipes', 'Arundo donax', 'Rumex conglomeratus', 'Typha angustifolia']
```

```python
for i in wet:
    df.loc[df['Species name'] == i, 'aquatic'] = 1
```

# Nativity

```python
native = ['Urochloa arizonica', 'Eriogonum fasciculatum', 'Eriogonum fasciculatum', 'Talinum paniculatum', 
          'Dysphania ambrosioides', 'Gamochaeta purpurea', 'Amaranthus palmeri', 'Elymus elymoides', 
          'Verbesina encelioides', 'Euphorbia davidii', 'Portulaca oleracea']
```

```python
for i in native:
    df.loc[df['Species name'] == i, 'nativity_sw'] = 'native'
```

```python
df.loc[df['Species name'] =='Solanum americanum', 'nativity_sw'] = 'non-native'
```

```python
df[(df.aquatic != 1) & (df.duration_consensus != 'annual') & (df.nativity_sw != 'non-native') & (df.low_elev_m < 1000)].shape
```

```python
df.to_csv('plant_features.csv')
```
