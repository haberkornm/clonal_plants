```python
df['stolon'] = 0

for i in stolon:
    df.loc[df['Species name'] == i, 'stolon'] = 1
```

```python
df['Creeping'] = 0
```

```python
rooting = ['Euphorbia albomarginata', 'Echinochloa colona', 'Eriochloa aristata', 'Eriochloa acuminata', 'Erythranthe guttata', 
           'Poa annua', 'Glandularia bipinnatifida', 'Nasturtium officinale', 'Oxalis albicans', 'Bouteloua repens', 'Chamaesyce albomarginata', 
           'Melinis repens', 'Oxalis corniculata', 'Lophocereus schottii', 'Eragrostis lehmanniana', 'Commelina erecta', 'Melinis repens']
```

```python
for i in rooting:
    df.loc[df.Species == i, 'Creeping'] = 1
```

```python
df.loc[df.Genus == "Agave", 'rhizome'] = 1
```

```python
bulb = ['Anemone tuberosa', 'Allium macropetalum', 'Zephyranthes longifolia', 'Triteleiopsis palmeri', 'Dichelostemma capitatum']
```

```python
df['bulb split'] = 0
for i in bulb:
    df.loc[df['Species name'] == i, 'bulb split'] = 0
```

```python
df.loc[df.Genus == 'Lycium', 'rhizome'] = 1
```

```python
df['root budding'] = 0
df.loc[df.Genus == 'Salix', 'root budding'] = 1
df.loc[df['Species name'] == 'Hoffmannseggia glauca', 'root budding'] = 1
```

```python
df.loc[df.Family == 'TYPHACEAE', 'rhizome'] = 1
df.loc[df.Family == 'JUNCACEAE', 'rhizome'] = 1
df.loc[df.Family == 'CYPERACEAE', 'rhizome'] = 1
```

```python
df['tillering'] = 0
df.loc[df.Family == 'POACEAE', 'tillering'] = 1
```

```python
df = df[df.low_elev_m <= 1000].reset_index(drop = True)
df.shape
```

# Clonal plant data

```python
clones = pd.read_csv('Clones.csv')
clones = pd.DataFrame(clones[:118])
clones.rename(columns = {'Unnamed: 0' : 'Species name'}, inplace = True)
clones.shape
```

```python
int_cols = ['Root budding', 'Rhizome', 'Axial splitting', 'Creeping', 'Stolon', 'Bulb/corm', 'Stem budding',
            'Multiple CGO', 'Disturbance related', 'Psuedo annual', 'Connection >2 years', 'Connection<2 years', 
            'Spread >0.5M', 'Spread<0.5M', 'Multiplication frequent', 'Multiplication infrequent', 'Genet >5 ramets', 
            'Genet <=5 ramets', 'CGO water storage', 'Descendents smaller', 'Descendents same size']
```

```python
clones[int_cols] = clones[int_cols].astype(int)
```

```python
clones.info()
```

```python
clones['Life form.1'].value_counts()
```

```python
df.lifeform.value_counts()
```

```python
clones.loc[clones['Species name'] == 'Acacia greggii', 'Species name'] = 'Senegalia greggii'
clones.loc[clones['Species name'] == 'Aristida pupurea', 'Species name'] = 'Aristida purpurea'
clones.loc[clones['Species name'] == 'Cylinropuntia bigelovii', 'Species name'] = 'Cylindropuntia bigelovii'
clones.loc[clones['Species name'] == 'Digiteria californica', 'Species name'] = 'Digitaria californica'
clones.loc[clones['Species name'] == 'Hyptis emoryi', 'Species name'] = 'Condea emoryi'
clones.loc[clones['Species name'] == 'Janusia gracilis', 'Species name'] = 'Cottsia gracilis'
clones.loc[clones['Species name'] == 'Lycium exertum', 'Species name'] = 'Lycium exsertum'
clones.loc[clones['Species name'] == 'Lycium macrdon', 'Species name'] = 'Lycium macrodon'
clones.loc[clones['Species name'] == 'Oenothera californica', 'Species name'] = 'Oenothera arizonica'
clones.loc[clones['Species name'] == 'Trichloris crinita', 'Species name'] = 'Leptochloa crinita'
```

```python
clones = clones[clones['Species name'] != 'Lycium richii']
clones = clones[clones['Species name'] != 'Lycium cooperi']
```

```python
set(clones['Species name']) - set(df['Species name'])
```

```python
df = df.merge(clones, left_on = 'Species name', right_on = 'Species name', how = 'left')
```

```python
df.fillna(0, inplace = True)
```

```python
df.loc[(df['stolon'] == 0) & (df['Stolon'] == 1), 'stolon'] = 1
df.drop(columns = 'Stolon', inplace = True)
```

```python
df.loc[(df['rhizome'] == 0) & (df['Rhizome'] == 1), 'rhizome'] = 1
df.drop(columns = 'Rhizome', inplace = True)
```

```python
df.loc[(df['Creeping_x'] == 0) & (df['Creeping_y'] == 1), 'Creeping_x'] = 1
df.drop(columns = 'Creeping_y', inplace = True)
df.rename(columns = {'Creeping_x' : 'creeping'}, inplace = True)
```

```python
df.loc[(df['Bulb/corm'] == 0) & (df['bulb split'] == 1), 'Bulb/corm'] = 1
df.drop(columns = 'bulb split', inplace = True)
```

```python
df.loc[(df['root budding'] == 0) & (df['Root budding'] == 1), 'root budding'] = 1
df.drop(columns = 'Root budding', inplace = True)
```
